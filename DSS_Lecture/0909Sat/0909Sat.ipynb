```python
import numpy as np
import matplotlib as mpl
import matplotlib.pylab as plt
def f(x):
    return x**3 - 3*x**2 + x
```

```python
x = np.linspace(-1,3,9)
x
```

```python
y = f(x)
y
```

```python
plt.plot(x,y, 'ro-')
plt.xlim(-2, 4)
plt.xticks(np.arange(-1, 4))
plt.yticks(np.arange(-5, 4))

plt.show()
```

```python
x = np.linspace(-1,3,400)
y = f(x)
plt.plot(x,y)
plt.xlim(-2, 4)
plt.xticks(np.arange(-1, 4))
plt.yticks(np.arange(-5, 4))

plt.show()
```

```python
def f_exp(x):
    return np.e ** x

xx = np.linspace(-2, 2, 100)
yy = f_exp(xx)
plt.plot(xx, yy)
plt.axhline(1, c='r', ls="--")
plt.axvline(0, c='r', ls="--")
plt.show()
```
