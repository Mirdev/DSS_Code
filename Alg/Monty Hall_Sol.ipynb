```python
#Valid! and Standard of validation
#High-level lang ver.

#import some functions
from random import seed
from random import random
from time import time

#variable declare
n = 100000000
i = 0
win_nc = 0
win_c = 0
seed()
ts_t = time()

while i < n :
    #initialize in function
    door = [0,0,0]
    reveal = [0,1,2]
    select = int(random()*3)
    door[select] = 1 #car

    #Case 1 : Take a first_choice and don't change(do not take second_choice) + store the result
    choice = int(random()*3)
    win_nc += door[choice]

    #Case 2 : Open the 'fail' door and then change to second_choice + store the result
    
    ##Open the 'fail' door except choiced door
    reveal.remove(choice)
    
    ##Supervisor knows where is the car, so he also except the 'win' door
    ##If choice == select all remainders are 'fail' door. So we can remove any of them.
    if choice == select:
        reveal.pop(int(random()*2))
        ##We can also use 'reveal.pop(0)'. Because all remainders are 'fail' door. So we can remove any of them.
    else:
        reveal.remove(select)
    ##Now reveal[0] is storing revealed door index
    
    ##In this case, player have to change his selection.
    ##It means excepting revealed door, 'choice' value must have other index. And then store the result.
    for j in range(3):
        if j != choice and j != reveal[0]:
            win_c += door[j]
            break
    ##We can use 'win_c += door[3^choice^reveal[0]]' construct instead of above iteration.

    i += 1
    
#time check
print("Taking time :",time()-ts_t)

#result print
print(win_nc/n*100, "% ", win_c/n*100, "%")
```

```python
#Valid!
#Logical operation with bit operation ver Conditional statement

#import some functions
from random import seed
from random import random
from time import time

#variable declare
n = 100000000
i = 0
win_nc = 0
win_c = 0
seed()
ts_t = time()

while i < n :
    #initialize in function
    door = [0,0,0]
    select = int(random()*3)
    door[select] = 1 #car

    #take a first_choice and don't change(do not take second_choice) + count stack
    choice = int(random()*3)
    win_nc += door[choice]

    #Open the 'fail' door and then change to second_choice + count stack
    if select==choice:
        rvl = select
        if not select:
            rvl+=1
    else:
        rvl = select^choice
    win_c += door[choice^rvl]

    i += 1
    
print("Taking time :",time()-ts_t)
print(win_nc/n*100, "% ", win_c/n*100, "%")
```

```python
#Valid!
#Logically programmed with human brain
#It contains lots of assumes

from random import seed
from random import random
from time import time

#variable declare and initialize
n = 100000000
i = 0
win_nc = 0
win_c = 0
seed()
ts_t = time()

while i < n :
    #initialize in function
    door = [0,0,0]
    
    #set a car
    door[int(random()*3)] = 1

    #take a first_choice and don't change(do not take second_choice) + count stack
    choice = int(random()*3)
    win_nc += door[choice]

    #Open the 'fail' door and then change to second_choice + count stack
    ##We can imagine that changing first choice means opposite('not') result of first choice even open 'fail' door
    ##And also change the door is same to change result
    ##We can refer 'https://en.wikipedia.org/wiki/Monty_Hall_problem#Simple_solutions'
    win_c += not door[choice]

    i += 1

#time check
print("Taking time :",time()-ts_t)

#result print
print(win_nc/n*100, "% ", win_c/n*100, "%")
```
